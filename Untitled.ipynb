```python
final = []
run = 0

run+=1
skew = 5
width = 0.1
print(run)

# get IVs, set index to 'Date' & sort 

# sample = pd.read_csv('https://data.nasdaq.com/api/v3/datasets/VOL/MSFT.csv?api_key=T2VW5wXL4RvSbrytEKuD')


sample_data = sample[['Date', 'IvCall10', 'IvPut10']].copy()
sample_data.index = pd.to_datetime(sample_data.Date)
sample_data.drop('Date', inplace = True, axis=1)
sample_data.sort_index(inplace=True)

# get min and max date 

max_date = sample_data.index.max()
min_date = sample_data.index.min()

# download price & rate 

stock  = yf.download('MSFT')
rate = yf.download('^IRX')['Adj Close']/100

# set index to datetime & get price in range of dates

stock.index = pd.to_datetime(stock.index)
stock = stock.loc[(stock.index >= min_date) & (stock.index <= max_date)]
rate = rate.loc[(rate.index >= min_date) & (rate.index <= max_date)]

sclose = stock['Close']
sopen = stock['Open']

# set Price to price and use only dates from rate, then set rate to rate

sample_data['Close'] = sclose
sample_data['Open'] = sopen
sample_data = sample_data[sample_data.index.isin(list(rate.index))]
sample_data['Rate'] = rate

# Binomial pricign model 

def Binomial(S, K, r, v, t, PutCall):  
    n = 100
    At = t/n 
    u = np.exp(v*np.sqrt(At))
    d = 1./u
    p = (np.exp(r*At)-d) / (u-d) 

    #Binomial price tree
    stockvalue = np.zeros((n+1,n+1))
    stockvalue[0,0] = S
    for i in range(1,n+1):
        stockvalue[i,0] = stockvalue[i-1,0]*u
        for j in range(1,i+1):
            stockvalue[i,j] = stockvalue[i-1,j-1]*d

    #option value at final node   
    optionvalue = np.zeros((n+1,n+1))
    for j in range(n+1):
        if PutCall=="C": # Call
            optionvalue[n,j] = max(0, stockvalue[n,j]-K)
        elif PutCall=="P": #Put
            optionvalue[n,j] = max(0, K-stockvalue[n,j])

    #backward calculation for option price    
    for i in range(n-1,-1,-1):
        for j in range(i+1):
                if PutCall=="P":
                    optionvalue[i,j] = max(0, K-stockvalue[i,j], np.exp(-r*At)*(p*optionvalue[i+1,j]+(1-p)*optionvalue[i+1,j+1]))
                elif PutCall=="C":
                    optionvalue[i,j] = max(0, stockvalue[i,j]-K, np.exp(-r*At)*(p*optionvalue[i+1,j]+(1-p)*optionvalue[i+1,j+1]))
    return optionvalue[0,0]


# get days til next friday & if its less than 4 days use the friday after so 4 <= t <= 10


min_days -= 1
sample_data['days til friday'] = 4 - sample_data.index.weekday + 7 
sample_data['days til next friday'] = sample_data['days til friday'] + 14
sample_data['chose next'] =  (sample_data['days til friday'] <= min_days).astype(int)
sample_data['chose this'] =  (sample_data['days til friday'] > min_days).astype(int)
sample_data['t'] = (sample_data['days til next friday'] * sample_data['chose next']) +(sample_data['days til friday'] * sample_data['chose this'])
sample_data.drop(['days til friday','days til next friday', 'chose next', 'chose this'], inplace=True, axis=1)

# change data name to df

df = sample_data

# get mean and stdev to estimate price for strike of the mid call

df['mean return'] = np.log(df['Close'].shift(1)/df['Close'].shift(2)).rolling(lookback).mean()
df['mean std'] = np.log(df['Close'].shift(1)/df['Close'].shift(2)).rolling(lookback).std()
df.dropna(axis=0, inplace=True)


def get_estimate(x):

    final_returns = (x[6] + x[7] * np.random.normal(size=(int(1000),int(x[5])))).tolist()

    final_values = []
    for i in range(len(final_returns)):
        a = final_returns[i]
        values = []
        values.append(x[2])
        for i in range(len(a)):
            values.append((values[i]*(1+a[i])))
        final_values.append(values)

    values_df = pd.DataFrame(final_values).transpose()    
    last_days = values_df.iloc[-1].tolist()
    last_days_rounded = [round(num, 0) for num in last_days]
    return last_days_rounded

# Close = yesterdays close 

df['Close'] = df['Close'].shift(1)
df.dropna(axis=0, inplace=True)

# apply get_estimate to get estimated future prices 

print('estimating prices...')

df['Price Estimate'] = df.apply(get_estimate, axis=1)
df['K2'] = round(df['Price Estimate'].apply(np.mean),0)+skew
df['spread'] = round(df['K2']*width, 0)                  #width gets used here for the Ks
df['K1'] = round(df['K2']-df['spread'], 0)     
df['K3'] = round(df['K2']+df['spread'], 0)

# get call prices for each day 

def call_1(x):
    return Binomial(x[3], x[11], x[4], x[0], x[5]/365, 'C')* 1.02

def call_2(x):
    return Binomial(x[3], x[9], x[4], x[0], x[5]/365, 'C')* 0.98

def call_3(x):
    return Binomial(x[3], x[12], x[4], x[0], x[5]/365, 'C')* 1.02

print('Calculating Call Prices...')

df['C1'] = sample_data.apply(call_1, axis=1) 
df['C2'] = sample_data.apply(call_2, axis=1) 
df['C3'] = sample_data.apply(call_3, axis=1) 

# get payout pattern for each day 

def get_payout_1(x):
    payout = pd.DataFrame(list(range((0), int(x[2]*4))), columns = ['Price'])
    payout['Zero'] = 0
    payout['Diff'] = payout['Price'] -x[11]
    payout['Payout'] = payout[['Zero', 'Diff']].max(axis=1) - x[13]
    payout.index = payout['Price']
    payout.drop(['Zero', 'Diff', 'Price'], inplace=True, axis=1)
    return payout

def get_payout_2(x):
    payout = pd.DataFrame(list(range((0), int(x[2]*4))), columns = ['Price'])
    payout['Zero'] = 0
    payout['Diff'] = payout['Price'] -  x[9]
    payout['Payout'] = payout[['Zero', 'Diff']].max(axis=1) - x[14]
    payout.index = payout['Price']
    payout.drop(['Zero', 'Diff', 'Price'], inplace=True, axis=1)
    return payout

def get_payout_3(x):
    payout = pd.DataFrame(list(range((0), int(x[2]*4))), columns = ['Price'])
    payout['Zero'] = 0
    payout['Diff'] = payout['Price'] -  x[12]
    payout['Payout'] = payout[['Zero', 'Diff']].max(axis=1) - x[15]
    payout.index = payout['Price']
    payout.drop(['Zero', 'Diff', 'Price'], inplace=True, axis=1)
    return payout

print('Calculating Payouts...')

df['Payout 1'] = df.apply(get_payout_1, axis=1)
df['Payout 2'] = df.apply(get_payout_2, axis=1)
df['Payout 3'] = df.apply(get_payout_3, axis=1)
df['Total Payout'] = 100*df['Payout 1'] -200* df['Payout 2'] + 100*df['Payout 3']


# estimate prices with different lookbacks 

print('Calculating Expected Values...')

def get_ev(x):
    return np.sum([[x[19].loc[int(i)] for i in x[8]]])/1000

a = 1
for lookback in [5,20,40,120]:
    df['mean return'] = np.log(df['Close'].shift(1)/df['Close'].shift(2)).rolling(lookback).mean()
    df['mean std'] = np.log(df['Close'].shift(1)/df['Close'].shift(2)).rolling(lookback).std()
    df.dropna(inplace=True)
    
    df['Price Estimate'] = df.apply(get_estimate, axis=1)
    df['Expected Value '+str(a)] = df.apply(get_ev, axis=1)
    
    a+=1 

df['Average Expected Value'] = (df['Expected Value 1'] + df['Expected Value 2'] + df['Expected Value 3'] + df['Expected Value 4'])/4
# calculate the expected value by multiplying the payouts by the ecpected prices 


# df['Expected Values'] = df.apply(get_ev, axis=1)
# tries = []
# tries.append(np.mean(df['Expected Values']))
# tries.append(width)
# tries.append(skew)
# tries.append(lookback)
# final.append(tries)

# pd.DataFrame(final).to_excel('test2.xlsx')
```
