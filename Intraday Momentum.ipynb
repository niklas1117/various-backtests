```python
import yfinance as yf
import pandas as pd 
import datetime as dt 
```

```python
stock = 'AAPL'
```

```python
intraday = yf.download(stock, start = '2021-03-25', end= '2021-03-26', interval = '1m')
```

```python
intraday.iloc[120].Open/intraday.iloc[0].Open -1
```

```python
def Intradaytrend(df, entry, exit):
    ret_120min = df.iloc[120].Open/df.iloc[0].Open -1
    tickret = df.Open.pct_change()
    if ret_120min > entry:
        buyprice = df.iloc[121].Open
        buytime = df.iloc[121].name
        cumulated = (tickret.loc[buytime:] + 1).cumprod()-1
        exittime = cumulated[(cumulated < -exit) | (cumulated > exit)].first_valid_index()
        if exittime == None:
            exitprice = df.iloc[-1].Open
        else:
            exitprice = df.loc[exittime + dt.timedelta(minutes = 1)].Open
        profit = exitprice - buyprice
        profitrel = profit/buyprice
        return profitrel
    else:
        return None 
```

```python
Intradaytrend(intraday, 0.01, 0.01)
```

```python
stock = 'AMD'
dates = yf.download(stock, start= '2021-03-07', end='2021-04-07')
frame = []
for i in dates.index:
    frame.append(yf.download(stock, start= i, end=i + dt.timedelta(days =1), interval = '1m'))
```

```python
del frame[0:1]
```

```python
returns = []
for i in frame:
    returns.append(Intradaytrend(i, 0.01, 0.02))
```
